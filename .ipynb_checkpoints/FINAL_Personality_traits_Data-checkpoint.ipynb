```python
dataset['gender'] = dataset['gender'].replace(0, method = 'bfill')
```

```python
dataset = dataset.replace(0, method = 'bfill')
```

```python
dataset.head()
```

```python
# Get location of the last column to include in the X values, gender is the Y value
x_last_column = dataset.columns.get_loc('P10')
```

```python
X = dataset.loc[:, 'A1':'P10'].values
Y = dataset['gender'].values
```

```python
X.shape
```

```python
Y.shape
```

```python
import xgboost as xgb
```

```python
random_seed = 42
test_size = 0.30


X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=test_size, random_state=random_seed)
```

```python
#D_train = xgb.DMatrix(X_train, label=Y_train)
#D_test = xgb.DMatrix(X_test, label=Y_test)
```

```python
# Fit Model
model = XGBClassifier()
model.fit(X_train, y_train)
```

```python
# Make preditions for test data
y_pred = model.predict(X_test)
predictions = [round(value) for value in y_pred]
```

```python
# evaluate predictions
accuracy = accuracy_score(y_test, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```

```python
print('Test Accuracy', model.score(X_test, y_pred))
```
